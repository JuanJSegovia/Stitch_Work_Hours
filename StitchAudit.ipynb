```python
df_dedup = df.drop_duplicates(subset=['Date', 'Time', 'User']) # Deduplicate based on user and date/time / Remove financial services cloud
```

```python
df_dedup['User'].value_counts() # AFTER deduplication
```

```python
df_dedup['actions_per_day'] = df_dedup.groupby(['Date'])['Time'].transform('size') # Added a new column to find out actions per day
```

```python
df_dedup = df_dedup.loc[df_dedup.actions_per_day > 1] # Dropping 7 days with only 1 login
df_dedup.sample(2)
```

```python
## CREATE A NEW TIMESTAMP COLUMN 

# Make sure Date is datetime64
df_dedup["Date"] = pd.to_datetime(df_dedup["Date"], errors="coerce")

# Clean Time: normalize whitespace + remove trailing timezone abbreviation (PST/PDT/etc)
time_clean = (
df_dedup["Time"].astype(str)
.str.replace(r"\s+", " ", regex=True) # collapses weird/multiple spaces
.str.strip()
.str.replace(r"\s+[A-Z]{2,4}$", "", regex=True) # drops final token like PST/PDT/UTC
)

# Build combined datetime strings
dt_str = df_dedup["Date"].dt.strftime("%Y-%m-%d") + " " + time_clean

# Parse with seconds, then fallback without seconds
ts1 = pd.to_datetime(dt_str, format="%Y-%m-%d %I:%M:%S %p", errors="coerce")
ts2 = pd.to_datetime(dt_str, format="%Y-%m-%d %I:%M %p", errors="coerce")

df_dedup["timestamp"] = ts1.fillna(ts2)
```

```python
df_dedup.sample(3)
```

```python
# Min/max per user per day + total time between
g = df_dedup.groupby(["Date", "User"])["timestamp"]
df_dedup["min_login_dt"] = g.transform("min")
df_dedup["max_login_dt"] = g.transform("max")
df_dedup["total_time_per_day"] = df_dedup["max_login_dt"] - df_dedup["min_login_dt"] # Timedelta

# Optional numeric versions
df_dedup["total_time_per_day_minutes"] = df_dedup["total_time_per_day"].dt.total_seconds() / 60
df_dedup["total_time_per_day_hours"] = df_dedup["total_time_per_day"].dt.total_seconds() / 3600
```

```python
df_dedup.sample(2)
```

```python
df_dedup2 = df_dedup.drop_duplicates(subset=['Date', 'User'], keep='first') # Deduplicate based on user and date aggregation
```

```python
df_dedup2.groupby('User')['total_time_per_day_hours'].sum() # TOTAL HOURS WORKED ACROSS INDIVIDUAL ACCOUNTS
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(df_dedup2['Date'], df_dedup2['actions_per_day'])
plt.xlabel('Date')
plt.ylabel('Total Actions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
